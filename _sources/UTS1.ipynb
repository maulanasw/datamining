```python
#Pisahkan dataset menjadi 80% set Pelatihan dan 20% set Pengujian
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 0)
```

```python
#Penskalaan Fitur
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train[:5]
```

```python
def models(X_train,Y_train):
  
  #Menggunakan Metode KNeighborsClassifier dari kelas tetangga untuk menggunakan algoritma Nearest Neighbor
  from sklearn.neighbors import KNeighborsClassifier
  knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
  knn.fit(X_train, Y_train)

    
  #mencetak akurasi model pada data pelatihan.
  print(knn.score(X_train, Y_train))
  return knn

model = models(X_train,Y_train)
```
